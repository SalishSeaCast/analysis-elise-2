```python
ii=~pd.isna(df1.A_diam_microns)
A_diam=df1.loc[ii,['A_diam_microns']].values[:,0]
A_count=df1.loc[ii,['ACountsPerML']].values[:,0]
ii=~pd.isna(df1.C_diam_microns)
C_diam=df1.loc[ii,['C_diam_microns']].values[:,0]
C_count=df1.loc[ii,['CCountsPerML']].values[:,0]
```

```python
plt.plot(A_diam,A_count/np.sum(A_count),'r.')
plt.plot(C_diam,C_count/np.sum(C_count),'b.')
```

```python
plt.plot(np.log10(A_diam),A_count/np.sum(A_count),'r.')
plt.plot(np.log10(C_diam),C_count/np.sum(C_count),'b.')
```
