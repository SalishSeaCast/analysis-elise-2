```python
import matplotlib.pyplot as plt
import netCDF4 as nc
import numpy as np
%matplotlib inline
```

```python
f=nc.Dataset('/results/SalishSea/spinup.201905/05oct13/SalishSea_1h_20131005_20131005_ptrc_T.nc')
```

```python
fig,ax=plt.subplots(1,3,figsize=(12,6))
m1=ax[0].pcolormesh(f.variables['diatoms'][0,0,:,:])
fig.colorbar(m1,ax=ax[0])
m1=ax[1].pcolormesh(f.variables['silicon'][0,0,:,:])
fig.colorbar(m1,ax=ax[1])
m1=ax[2].pcolormesh(f.variables['nitrate'][0,0,:,:])
fig.colorbar(m1,ax=ax[2])
```

```python
fig,ax=plt.subplots(3,1,figsize=(12,6))
m1=ax[0].pcolormesh(np.flipud(f.variables['diatoms'][0,:,500,200:]))
fig.colorbar(m1,ax=ax[0])
m1=ax[1].pcolormesh(np.flipud(f.variables['silicon'][0,:,500,200:]))
fig.colorbar(m1,ax=ax[1])
m1=ax[2].pcolormesh(np.flipud(f.variables['nitrate'][0,:,500,200:]))
fig.colorbar(m1,ax=ax[2])
```

```python
fig,ax=plt.subplots(3,1,figsize=(12,6))
m1=ax[0].pcolormesh(np.flipud(f.variables['particulate_organic_nitrogen'][0,:,500,200:]))
fig.colorbar(m1,ax=ax[0])
m1=ax[1].pcolormesh(np.flipud(f.variables['dissolved_organic_nitrogen'][0,:,500,200:]))
fig.colorbar(m1,ax=ax[1])
m1=ax[2].pcolormesh(np.flipud(f.variables['biogenic_silicon'][0,:,500,200:]),vmin=0,vmax=.05)
fig.colorbar(m1,ax=ax[2])
```
