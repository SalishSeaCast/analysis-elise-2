```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def Ilim(vm,alpha,I,theta):
    return vm*(1-np.exp(-1*alpha*I*theta/vm))
```

```python
I=np.arange(0,150)
```

```python
plt.plot(I,Ilim(2.2,3,I,.0535),'b-',label='diatoms')
plt.plot(I,Ilim(1.8,1.5,I,.0535),'r-',label='flagellates')
plt.xlabel('I')
plt.ylabel('growth rate')
plt.legend
```

```python
def Nlim(vm,NO3,NH4,kn,kr):
    return vm*(NO3/(kn+NO3)*kr/(kr+NH4)+NH4/(kr+NH4))
```

```python
NO3=np.arange(0,35)
NH4=np.arange(0,8,.2)
kn1=.75; kn2=1.0
kr1=.5;kr2=.75
vm1=1.8;vm2=2.2
fig,ax=plt.subplots(1,2,figsize=(12,5))
ax[0].plot(NO3,Nlim(vm1,NO3,0,kn1,kr1),'r-',label='flags NH4=0')
ax[0].plot(NO3,Nlim(vm1,NO3,2,kn1,kr1),'r--',label='flags NH4=2')
ax[0].plot(NO3,Nlim(vm1,NO3,4,kn1,kr1),'r:',label='flags NH4=4')
ax[0].plot(NO3,Nlim(vm2,NO3,0,kn2,kr2),'b-',label='diat NH4=0')
ax[0].plot(NO3,Nlim(vm2,NO3,2,kn2,kr2),'b--',label='diat NH4=2')
ax[0].plot(NO3,Nlim(vm2,NO3,4,kn2,kr2),'b:',label='diat NH4=4')
ax[0].legend()
ax[0].set_xlabel('NO3')
ax[0].set_ylabel('N limited growth')
ax[1].plot(NH4,Nlim(vm1,0,NH4,kn1,kr1),'r-',label='flags NO3=0')
ax[1].plot(NH4,Nlim(vm1,2,NH4,kn1,kr1),'r--',label='flags NO3=2')
ax[1].plot(NH4,Nlim(vm1,10,NH4,kn1,kr1),'r:',label='flags NO3=10')
ax[1].plot(NH4,Nlim(vm2,0,NH4,kn2,kr2),'b-',label='diat NO3=0')
ax[1].plot(NH4,Nlim(vm2,2,NH4,kn2,kr2),'b--',label='diat NO3=2')
ax[1].plot(NH4,Nlim(vm2,10,NH4,kn2,kr2),'b:',label='diat NO3=10')
ax[1].legend()
ax[1].set_xlabel('NH4')
ax[1].set_ylabel('N limited growth')
```

```python
def IlimSSC(mu,I,Iopt):
    # make relative to 10C like Pena
    return mu*np.exp(.07*(10-20))*1.06*np.exp(-I/(30*Iopt))*(1-np.exp(-I/(.33*Iopt)))
```

```python
# add ssc c-p2 values
```

```python
plt.plot(I,Ilim(2.2,3,I,.0535),'b-',label='diatoms')
plt.plot(I,Ilim(1.8,1.5,I,.0535),'r-',label='flagellates')
mudiat=6.11E-05*24*3600;muflag=2.11E-05*24*3600
Ioptdiat=42;Ioptflag=10
plt.plot(I,IlimSSC(mudiat,I,Ioptdiat),'c-',label='SSC diatoms')
plt.plot(I,IlimSSC(6.0495E-05*24*3600,I,45),'g-',label='SSC1905 diatoms')
plt.plot(I,IlimSSC(muflag,I,Ioptflag),'m-',label='SSC flagellates')
plt.xlabel('I')
plt.ylabel('growth rate')
plt.legend()
```

```python
def NlimSSC(mu,kappa,k,NO3,NH4):
    # make relative to 10C like Pena
    return mu*np.exp(.07*(10-20))*(kappa*NO3/(k+kappa*NO3+NH4)+NH4/(k+kappa*NO3+NH4))
```

```python
# add SSC 201812 values
NO3=np.arange(0,35)
NH4=np.arange(0,8,.2)
kn1=.75; kn2=1.0
kr1=.5;kr2=.75
vm1=1.8;vm2=2.2
kapdiat=1;kapflag=.3
kdiat=2;kflag=.1
mudiat=6.11E-05*24*3600;muflag=2.11E-05*24*3600
fig,ax=plt.subplots(1,2,figsize=(12,5))
ax[0].plot(NO3,Nlim(vm1,NO3,0,kn1,kr1),'r-',label='flags NH4=0')
ax[0].plot(NO3,Nlim(vm1,NO3,2,kn1,kr1),'r--',label='flags NH4=2')
ax[0].plot(NO3,Nlim(vm1,NO3,4,kn1,kr1),'r:',label='flags NH4=4')
ax[0].plot(NO3,Nlim(vm2,NO3,0,kn2,kr2),'b-',label='diat NH4=0')
ax[0].plot(NO3,Nlim(vm2,NO3,2,kn2,kr2),'b--',label='diat NH4=2')
ax[0].plot(NO3,Nlim(vm2,NO3,4,kn2,kr2),'b:',label='diat NH4=4')
ax[0].plot(NO3,NlimSSC(muflag,kapflag,kflag,NO3,0),'m-',label='SSC flags NH4=0')
ax[0].plot(NO3,NlimSSC(muflag,kapflag,kflag,NO3,2),'m--',label='SSC flags NH4=2')
ax[0].plot(NO3,NlimSSC(muflag,kapflag,kflag,NO3,4),'m:',label='SSC flags NH4=4')
ax[0].plot(NO3,NlimSSC(mudiat,kapdiat,kdiat,NO3,0),'c-',label='SSC diat NH4=0')
ax[0].plot(NO3,NlimSSC(mudiat,kapdiat,kdiat,NO3,2),'c--',label='SSC diat NH4=2')
ax[0].plot(NO3,NlimSSC(mudiat,kapdiat,kdiat,NO3,4),'c:',label='SSC diat NH4=4')
ax[0].legend()
ax[0].set_xlabel('NO3')
ax[0].set_ylabel('N limited growth')
ax[1].plot(NH4,Nlim(vm1,0,NH4,kn1,kr1),'r-',label='flags NO3=0')
ax[1].plot(NH4,Nlim(vm1,2,NH4,kn1,kr1),'r--',label='flags NO3=2')
ax[1].plot(NH4,Nlim(vm1,10,NH4,kn1,kr1),'r:',label='flags NO3=10')
ax[1].plot(NH4,Nlim(vm2,0,NH4,kn2,kr2),'b-',label='diat NO3=0')
ax[1].plot(NH4,Nlim(vm2,2,NH4,kn2,kr2),'b--',label='diat NO3=2')
ax[1].plot(NH4,Nlim(vm2,10,NH4,kn2,kr2),'b:',label='diat NO3=10')
ax[1].plot(NH4,NlimSSC(muflag,kapflag,kflag,0,NH4),'m-',label='flags NO3=0')
ax[1].plot(NH4,NlimSSC(muflag,kapflag,kflag,2,NH4),'m--',label='flags NO3=2')
ax[1].plot(NH4,NlimSSC(muflag,kapflag,kflag,10,NH4),'m:',label='flags NO3=10')
ax[1].plot(NH4,NlimSSC(mudiat,kapdiat,kdiat,0,NH4),'c-',label='diat NO3=0')
ax[1].plot(NH4,NlimSSC(mudiat,kapdiat,kdiat,2,NH4),'c--',label='diat NO3=2')
ax[1].plot(NH4,NlimSSC(mudiat,kapdiat,kdiat,10,NH4),'c:',label='diat NO3=10')
ax[1].legend()
ax[1].set_xlabel('NH4')
ax[1].set_ylabel('N limited growth')
```
