## Light Response

```python
I=np.arange(0,200)
Iopt=42
L42=(1-np.exp(-I/(.33*Iopt)))*np.exp(-I/(30*Iopt))*1.06
Iopt=37.000
L37=(1-np.exp(-I/(.33*Iopt)))*np.exp(-I/(30*Iopt))*1.06
Iopt=10.000
L10=(1-np.exp(-I/(.33*Iopt)))*np.exp(-I/(30*Iopt))*1.06
Iopt=47.000
L47=(1-np.exp(-I/(.33*Iopt)))*np.exp(-I/(30*Iopt))*1.06
```

```python
plt.plot(I,L42,'b-',label='diatoms')
plt.plot(I,L37,'r-',label='M.rubrum')
plt.plot(I,L10,'c-',label='flags')
plt.plot(I,L47,'b--',label='47')
plt.legend()
plt.xlabel('W/m$^2$')
```

## Another formulation: Platt et al 1980
    - less restriction on ratio of alpha to beta
    - simplest way to code: add factor s in photoinhibition term

```python
Iopt=10.000
s=1
L10s1=(1-np.exp(-I/(.33*Iopt)))*np.exp(-I*s/(30*Iopt))*1.06
s=.5
L10sp5=(1-np.exp(-I/(.33*Iopt)))*np.exp(-I*s/(30*Iopt))*1.06
s=2
L10s2=(1-np.exp(-I/(.33*Iopt)))*np.exp(-I*s/(30*Iopt))*1.06
```

```python
plt.plot(I,L10sp5,'b-',label='flags s=.5')
plt.plot(I,L10s1,'k-',label='flags s=1')
plt.plot(I,L10s2,'c-',label='flags s=2')
plt.legend()
plt.xlabel('W/m$^2$')
```
