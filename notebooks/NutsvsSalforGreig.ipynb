```python
import numpy as np
import netCDF4 as nc
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
with nc.Dataset('/results/SalishSea/nowcast-green.201812/01jul17/SalishSea_1d_20170701_20170701_ptrc_T.nc') as f:
    fig,ax=plt.subplots(1,1,figsize=(3,5))
    ax.pcolormesh(f.variables['nitrate'][0,0,:,:])
```

```python
fig,ax=plt.subplots(4,5,figsize=(15,15))
ax=ax.flatten()
with nc.Dataset('/results/SalishSea/nowcast-green.201812/01jul17/SalishSea_1d_20170701_20170701_ptrc_T.nc') as f:
    with nc.Dataset('/results/SalishSea/nowcast-green.201812/01jul17/SalishSea_1d_20170701_20170701_grid_T.nc') as fT:
        for k in range(0,20):
            ax[k].plot(fT.variables['vosaline'][0,k,400:700,100:300],f.variables['nitrate'][0,k,400:700,100:300],'k.')
            ax[k].set_title('level'+str(k))
            ax[k].set_xlabel('S')
            ax[k].set_ylabel('NO3')
print('July 1')
```
