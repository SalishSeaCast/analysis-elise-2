```python
import numpy as np
import netCDF4 as nc
import matplotlib.pyplot as plt

%matplotlib inline
```

### differences on Jan 31

```python
d0='/home/sallen/202007/202007Gc_before_carb/'
d1='/home/sallen/202007/202007Gc_after_carb/'
```

```python
fname='SalishSea_1d_20150126_20150131_carp_T_20150131-20150131.nc'
```

```python
f0=nc.Dataset(d0+fname)
f1=nc.Dataset(d1+fname)
```

```python
f0.variables.keys()
```

```python
fig,ax=plt.subplots(1,2,figsize=(12,3))
m0=ax[0].pcolormesh(f0.variables['dissolved_inorganic_carbon'][0,0,:,:])
fig.colorbar(m0,ax=ax[0])
m1=ax[1].pcolormesh(f1.variables['dissolved_inorganic_carbon'][0,0,:,:]-f0.variables['dissolved_inorganic_carbon'][0,0,:,:])
fig.colorbar(m1,ax=ax[1])
ax[0].set_title('orig dic')
ax[1].set_title('difference')
```

```python
fig,ax=plt.subplots(1,2,figsize=(12,3))
m0=ax[0].pcolormesh(f0.variables['total_alkalinity'][0,0,:,:])
fig.colorbar(m0,ax=ax[0])
m1=ax[1].pcolormesh(f1.variables['total_alkalinity'][0,0,:,:]-f0.variables['total_alkalinity'][0,0,:,:])
fig.colorbar(m1,ax=ax[1])
ax[0].set_title('orig alk')
ax[1].set_title('difference')
```

```python
np.max(np.abs(f1.variables['dissolved_inorganic_carbon'][:,:,:,:]-f0.variables['dissolved_inorganic_carbon'][:,:,:,:]))
```

```python
np.max(np.abs(f1.variables['total_alkalinity'][:,:,:,:]-f0.variables['total_alkalinity'][:,:,:,:]))
```

```python
# check that there is a difference between alk and dic
plt.pcolor(f0.variables['total_alkalinity'][0,0,:,:]-f0.variables['dissolved_inorganic_carbon'][0,0,:,:])
plt.colorbar()
```
