```python
from salishsea_tools import evaltools as et
import datetime as dt
```

```python
start=dt.datetime(2014,11,1)
end=dt.datetime(2015,1,31)
d0='/home/sallen/202007/202007Gc_before_carb'
d1='/home/sallen/202007/202007Gc_after_carb'
fl0=et.index_model_files(start,end,d0,'long',flen=1,ftype='carp_T',tres=24)
fl1=et.index_model_files(start,end,d1,'long',flen=1,ftype='carp_T',tres=24)
```

```python
with nc.Dataset('/data/eolson/MEOPAR/NEMO-forcing-new/grid/mesh_mask201702.nc') as mesh:
    tmask=mesh.variables['tmask'][:,:,:,:]
```

```python
len(fl0),len(fl1)
```

```python
maxdifdic=list()
maxdifalk=list()
for ind in range(0,len(fl0)):
    f0=nc.Dataset(fl0.paths[ind])
    f1=nc.Dataset(fl1.paths[ind])
    maxdifalk.append(np.max(np.abs(tmask*f1.variables['total_alkalinity'][:]-tmask*f0.variables['total_alkalinity'][:])))
    maxdifdic.append(np.max(np.abs(tmask*f1.variables['dissolved_inorganic_carbon'][:]-tmask*f0.variables['dissolved_inorganic_carbon'][:])))
```

```python
fig,ax=plt.subplots(1,1,figsize=(18,3))
ax.plot(maxdifalk,'r*',label='alk')
ax.plot(maxdifdic,'k*',label='dic')
ax.set_xlabel('Day of run')
ax.set_title('max(abs(difference))')
ax.legend()
```
