```python
import numpy as np
from matplotlib import pyplot as plt
import netCDF4 as nc
import cmocean

%matplotlib inline
```

```python
f0=nc.Dataset('/data/sallen/results/MEOPAR/202007/try3/SalishSea_1d_20150301_20150305_ptrc_T_20150301-20150301.nc')
f1=nc.Dataset('/data/sallen/results/MEOPAR/202007/tvd_sink/SalishSea_1d_20150301_20150305_ptrc_T_20150301-20150301.nc')
```

```python
for var in ('diatoms','biogenic_silicon','particulate_organic_nitrogen','flagellates','nitrate','silicon'):
    fig,ax=plt.subplots(3,1,figsize=(15,5))
    ax[0].pcolormesh(np.arange(0,898),f0.variables['deptht'][:],f0.variables[var][0,:,:,250])
    ax[0].set_ylim(450,0)
    ax[0].set_title(var+' old')
    ax[1].pcolormesh(np.arange(0,898),f1.variables['deptht'][:],f1.variables[var][0,:,:,250])
    ax[1].set_ylim(450,0)
    ax[1].set_title(var+' tvd sink')
    m0=ax[2].pcolormesh(np.arange(0,898),f1.variables['deptht'][:],
                     f1.variables[var][0,:,:,250]-f0.variables[var][0,:,:,250],
                    cmap=cmocean.cm.balance,vmin=-.1,vmax=.1)
    fig.colorbar(m0,ax=ax[2])
    ax[2].set_ylim(450,0)
    ax[2].set_title(var+' tvd sink - old')
```
