```python
iit=24
fig,ax=plt.subplots(2,5,figsize=(22,16))
ax=ax.flatten()
tt=0
#for tt, var in enumerate(['TVDUPW', 'TVDAD', 'TVDNO', 'TRA0', 'TRAADV', 
#                          'TRALDF', 'TRBZDF', 'TRAZDF', 'bSi']):
ax[tt].set_title('TVDUPW: upwind \ncomponent of tra')
ax[tt].set_ylim(40,0)
ax[tt].plot((0,0),(0,40),'k-',lw=1)
for ii in range(iit,iit+3):
    ax[tt].plot(f.variables['TVDUPW'][ii,:,j1,i1],np.arange(0,40),'*',color=cols[ii-iit])
ax[tt].set_xlim(-5e-4,5e-4)
    
tt+=1
ax[tt].set_title('actual advective trend')
ax[tt].plot((0,0),(30,40),'k-',lw=1)
for ii in range(iit,iit+3):
    ax[tt].plot(f.variables['TRAADV'][ii,:,j1,i1]-f.variables['TRA0'][ii,:,j1,i1],np.arange(0,40),'*',color=cols[ii-iit])
ax[tt].set_ylim(40,0)
ax[tt].set_xlim(-1e-5,1e-5)

tt+=1
ax[tt].set_title('ZWZUPW: zwz \n upwind component')
ax[tt].set_ylim(40,0)
ax[tt].plot((0,0),(0,40),'k-',lw=1)
for ii in range(iit,iit+3):
    ax[tt].plot(f.variables['ZWZUPW'][ii,:,j1,i1],np.arange(0,40),'*',color=cols[ii-iit])
ax[tt].set_xlim(-20,20)
    
tt+=1
ax[tt].set_title('TVDAD: zwz \n antidiff component')
ax[tt].set_ylim(40,0)
ax[tt].plot((0,0),(0,40),'k-',lw=1)
for ii in range(iit,iit+3):
    ax[tt].plot(f.variables['TVDAD'][ii,:,j1,i1],np.arange(0,40),'*',color=cols[ii-iit])
ax[tt].set_xlim(-20,20)

tt+=1
ax[tt].set_title('TVDNO: zwz \n after nonosc')
ax[tt].set_ylim(40,0)
ax[tt].plot((0,0),(0,40),'k-',lw=1)
for ii in range(iit,iit+3):
    ax[tt].plot(f.variables['TVDNO'][ii,:,j1,i1],np.arange(0,40),'*',color=cols[ii-iit])
ax[tt].set_xlim(-20,20)

tt+=1
#non-bio w
ax[tt].set_title('PHYWVEL: phys only w')
ax[tt].set_ylim(40,0)
ax[tt].plot((0,0),(0,40),'k-',lw=1)
for ii in range(iit,iit+3):
    ax[tt].plot(f.variables['PHYWVEL'][ii,:,j1,i1],np.arange(0,40),'*',color=cols[ii-iit])
    ax[tt].plot(f.variables['BIOWVEL'][ii,:,j1,i1],np.arange(0,40),'*',color='dodgerblue')
#ax[tt].set_xlim(-20,20)

tt+=1
#full w BIOWVEL
ax[tt].set_title('PHYWVEL: full w \n (phys+bio)')
ax[tt].set_ylim(40,0)
ax[tt].plot((0,0),(0,40),'k-',lw=1)
for ii in range(iit,iit+3):
    ax[tt].plot(f.variables['BIOWVEL'][ii,:,j1,i1]+f.variables['PHYWVEL'][ii,:,j1,i1],np.arange(0,40),'*',color=cols[ii-iit])
#ax[tt].set_xlim(-20,20)

tt+=1
ax[tt].set_title('ZWZUPW: zwz diff')
var=f.variables['ZWZUPW'][iit:(iit+3),:,j1,i1]
temp=np.zeros([3,40])
temp=temp-var
temp[:,:-1]=temp[:,:-1]+var[:,1:]
ax[tt].set_ylim(40,0)
ax[tt].plot((0,0),(0,40),'k-',lw=1)
for ii in range(0,3):
    ax[tt].plot(temp[ii,:],np.arange(0,40),'*',color=cols[ii])
ax[tt].set_xlim(-20,20)

tt+=1
ax[tt].set_title('TVDAD: zwz diff')
var=f.variables['TVDAD'][iit:(iit+3),:,j1,i1]
temp=np.zeros([3,40])
temp=temp-var
temp[:,:-1]=temp[:,:-1]+var[:,1:]
ax[tt].set_ylim(40,0)
ax[tt].plot((0,0),(0,40),'k-',lw=1)
for ii in range(0,3):
    ax[tt].plot(temp[ii,:],np.arange(0,40),'*',color=cols[ii])
ax[tt].set_xlim(-20,20)

tt+=1
ax[tt].set_title('TVDNO: zwz diff')
var=f.variables['TVDNO'][iit:(iit+3),:,j1,i1]
temp=np.zeros([3,40])
temp=temp-var
temp[:,:-1]=temp[:,:-1]+var[:,1:]
ax[tt].set_ylim(40,0)
ax[tt].plot((0,0),(0,40),'k-',lw=1)
for ii in range(0,3):
    ax[tt].plot(temp[ii,:],np.arange(0,40),'*',color=cols[ii])
ax[tt].set_xlim(-20,20)
```
