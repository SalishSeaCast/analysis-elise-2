```python
mins=list()
iis=list()
maxs=list()
for i in range(0,len(fc.variables['PAR'][:,0,0,0])):
    iis.append(i)
    mins.append(np.min(tmask*fc.variables['PAR'][i,:,:,:]))
    maxs.append(np.max(tmask*fc.variables['PAR'][i,:,:,:]))
```

```python
minsP=dict()
maxsP=dict()
for var in ('nitrate',):#fp.variables.keys():
    if len(np.shape(fp.variables[var][:]))==4:
        minsP[var]=list()
        maxsP[var]=list()
        for i in range(0,len(fp.variables['nitrate'][:,0,0,0])):
            minsP[var].append(np.min(tmask*fp.variables[var][i,:,:,:]))
            maxsP[var].append(np.max(tmask*fp.variables[var][i,:,:,:]))
```

```python
fig,ax=plt.subplots(1,1,figsize=(18,3))
ax.plot(iis,maxs,'r.')
ax.plot(iis,mins,'b.')
ax.set_title('PAR')
```

```python
for var in ('nitrate',):#fp.variables.keys():
    if len(np.shape(fp.variables[var][:]))==4:
        fig,ax=plt.subplots(1,2,figsize=(18,3))
        ax[0].plot(iis,maxsP[var],'r.')
        ax[1].plot(iis,minsP[var],'b.')
        ax[0].set_title(var)
```
