```python
import netCDF4 as nc
import numpy as np
from matplotlib import pyplot as plt
import cmocean as cm
import datetime as dt
import glob
from salishsea_tools import viz_tools
%matplotlib inline
```

# Observed Bloom Times

### 2015: March 6-16 - Surface Plots

```python
fformatP='SalishSea_1h_%Y%m%d_%Y%m%d_ptrc_T.nc'
fformat1='%d%b%y/'
fformatC='SalishSea_1h_%Y%m%d_%Y%m%d_carp_T.nc'
f2015=dict()
f2015['ptrc']=dict()
f2015['carp']=dict()
for nn in range(1,17):
    idt0=dt.datetime(2015,3,nn)
    cdir=idt0.strftime(fformat1).lower()
    iffP=idt0.strftime(fformatP)
    iffC=idt0.strftime(fformatC)
    ffP='/results2/SalishSea/hindcast/'+cdir+iffP
    ffC='/results2/SalishSea/hindcast/'+cdir+iffC
    f2015['ptrc'][nn]=nc.Dataset(ffP)
    f2015['carp'][nn]=nc.Dataset(ffC)
```

```python
cmap = plt.get_cmap(cm.cm.algae)   #('nipy_spectral')
cmap.set_bad('burlywood')
with nc.Dataset('/ocean/eolson/MEOPAR/NEMO-forcing/grid/mesh_mask201702_noLPE.nc') as fm:
    tmask=np.copy(fm.variables['tmask'])
fig,axs = plt.subplots(3,6,figsize=(24,15))
axs=axs.flatten()
for nn in range(1,17):
    ax = axs[nn-1]
    viz_tools.set_aspect(ax)
    mesh0 = ax.pcolormesh(np.ma.masked_where(tmask[0,0,200:750,50:]==0,f2015['ptrc'][nn].variables['diatoms'][18,0,200:750,50:])+f2015['ptrc'][nn].variables['ciliates'][18,0,200:750,50:]+f2015['ptrc'][nn].variables['flagellates'][18,0,200:750,50:], cmap=cmap)
    ax.set_title('2015-3-'+str(nn), fontsize=16)
    cbar = fig.colorbar(mesh0,ax=ax)
    cbar.ax.tick_params(labelsize=14)
    mesh0.set_clim(vmin=0, vmax=15)
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    ax.set_xlim((50,348))
```
