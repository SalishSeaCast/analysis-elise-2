```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import datetime as dt
%matplotlib inline
```

```python
df=pd.read_csv('/ocean/eolson/MEOPAR/obs/STRATOGEM/biovol/biovol_time_series_jun2012_new.csv')
```

```python
df['dt']=[dt.datetime(r.Year,r.Month,r.Day) for i,r in df.iterrows()]
df['yd']=[(dt.datetime(r.Year,r.Month,r.Day)-dt.datetime(r.Year-1,12,31)).days for i,r in df.iterrows()]
```

```python
df['Year'].unique()
```

## Note: 
- included in "small phytos": sum(dinoflagellates)+flagellates+cryptophytes
- maybe should not include flagellates: likely heterotrophic

```python
fig,ax=plt.subplots(1,1,figsize=(10,3))
for year in df['Year'].unique():
    df2=df.loc[df.Year==year]
    d,=ax.plot(df2['yd'],df2['Diatoms_mm3L'],'b-',label='diatoms')
    s,=ax.plot(df2['yd'],df2['smallPhyto_mm3L'],'c-',label='small phytos')
    m,=ax.plot(df2['yd'],df2['Mesodinium rubrum_mm3L'],'r-',label='Mesodinium')
ax.set_yscale('log')
ax.set_xlabel('Year Day')
ax.set_ylabel('Biovolume (mm$^3$/L)')
plt.legend(handles=[d,s,m],bbox_to_anchor=[1.25,1])
```

```python
fig,ax=plt.subplots(1,1,figsize=(7,8))
for year in df['Year'].unique():
    df2=df.loc[df.Year==year]
    d,=ax.plot(df2['yd'],df2['Diatoms_mm3L'],'b-',label='diatoms')
    s,=ax.plot(df2['yd'],df2['smallPhyto_mm3L'],'c-',label='small phytos')
    m,=ax.plot(df2['yd'],df2['Mesodinium rubrum_mm3L'],'r-',label='Mesodinium')
ax.set_ylim([0,5e7])
ax.set_xlabel('Year Day')
ax.set_ylabel('Biovolume (mm$^3$/L)')
plt.legend(handles=[d,s,m],bbox_to_anchor=[1,1])
```
